# data visualization

## segment 1 : creating standard data graphics

```python
# import libraries
# numpy
import numpy as np
from numpy.random import randn

# pandas
import pandas as pd
from pandas import Series, DataFrame

# matplot
import matplotlib.pyplot as plt
from matplotlib import rcParams

# utils
import os
```

### creating a line chart from a list object

#### plotting a line chart
#### - using matplotlib

```python
# create x-axis and y-axis
x = range(1, 10) # including 1, excluding 10
y = [-1, 0, 3, 9, 2, 6, -3, 0, 1]

# plot a line chart
plt.plot(x, y)
```

#### - using pandas objects

```python
# prepare data to plot
# get current working directory and data path
cwd = os.getcwd()
mtcars = cwd + '/linkedin-data/mtcars.csv'

cars = pd.read_csv(mtcars)

# rename unnamed column
cars.rename(columns={'Unnamed: 0': 'car_name'},
			inplace=True)

# display data
cars.head()
```

```python
# select 'mpg' info to plot
mpg = cars['mpg']

# plot mpg
mpg.plot()
```

```python
# select 'cyl', 'wt' and 'mpg' to plot
cars_data = cars[['cyl', 'wt', 'mpg']]

# plot selected columns
cars_data.plot()
```

### creating bar charts

#### - from a list

```python
# plot a bar chart
plt.bar(x, y)
```

#### - from pandas object

```python
# plot a bar chart - vertical as default
cars_data.plot(kind='bar')
```

```python
# plot a bar chart - horizontal
cars_data.plot(kind='barh')
# plt.savefig('test_barh.png') # will get the figure in the current directory
plt.show()
```

### creating a pie chart

```python
# mock up data
z = [0.5, 2, 6, 0]
plt.pie(z)
# plt.savefig('test_z.png') # will get the figure in the current directory
plt.show()
```

## segment 2 : defining elements of a plot

```python
# initial config
# to tell that the plot will inline w/ the notebook
%matplotlib inline

# size 5 x 4
rcParams['figure.figsize'] = 5, 4
```

### define axes, ticks and grids

```python
# to initialize the figure
fig = plt.figure()

# store the figure to the variable and add axes to the graph
# a, b, c, d
# left, bottom, width, height
ax = fig.add_axes([.1, .1, .5, .5])

ax.plot(x, y)
```

```python
# to initialize the figure
fig_2 = plt.figure()
ax_2 = fig_2.add_axes([.1, .1, .5, .5])

print('x min : ', min(x), ',	max : ', max(x))
print('y min : ', min(y), ',	max : ', max(y))

# set the limit of axes
ax_2.set_xlim([1, 9])
ax_2.set_ylim([-3, 9])

# set ticks
ax_2.set_xticks([1, 3, 5, 7, 9])
ax_2.set_yticks([-4, -2, 0, 2, 4, 6, 8])

# add grid to the figure
ax_2.grid()

# plot
ax_2.plot(x, y)
```

### generating multiple plots in one figure w/ subplots

```python
# to initial the figure
fig_3 = plt.figure()
fig_3, (ax, ax_2) = plt.subplots(1, 2)

ax.plot(x)
ax_2.plot(x, y)
```

### customizing line styles and adding plot markers

```python
# mock up another dataset
x1 = range(1, 11)
y1 = range(0, 10)

# plot 2 line charts in the same figure
plt.plot(x, y)
plt.plot(x1, y1)
```

```python
# format the above 2 charts
# ds = drawing style, lw = line width
plt.plot(x, y,
			ds='steps',
			lw=5,
			marker='+',
			mew=10
)

# ls = line style
plt.plot(x1, y1,
			ls='--',
			lw=2,
			marker='*',
			mew=5
)
```

## segment 4 : creating lables and annotations

### labeling plot features

```python
# functional method
plt.bar(x, y)

# define x and y axes' label
plt.xlabel('this is x label')
plt.ylabel('this is y label')
```

```python
# pie chart
v_type = ['bike', 'motorbike', 'van', 'stroller']

plt.pie(z, labels=v_type)
plt.show()
```

```python
# object-oriented method
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

cars.mpg.plot()
ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_name,
					rotation=60,
					fontsize='medium'
)
ax.set_title('mile per gallon')

ax.set_xlabel('car names')
ax.set_ylabel('miles / gallon')
```

### adding legend to a plot

```python
# functional method
# pie chart
plt.pie(z)
plt.legend(v_type, loc='best')
plt.show()
```

```python
# object-oriented method
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

cars.mpg.plot()

# set number of ticks and label of them
ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_name,
					rotation=60,
					fontsize='medium'
)

# set the name of the chart
ax.set_title('miles per gallon')

# set label
ax.set_xlabel('car names')
ax.set_ylabel('miles / gallon')

# position legend description
ax.legend(loc='best')
```

### annotation a plot

```python
# get the instance value
cars.mpg.max()
```

```python
# plot a chart
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])

# plt.plot(cars.mpg)
cars.mpg.plot()

# set number of ticks and label of them
ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_name,
					rotation=60,
					fontsize='medium'
)

ax.set_title('miles per gallon')

# set label and its location
ax.set_xlabel('miles / gallon')
ax.set_ylabel('car names')

# position legend description
ax.legend(loc='best')

# set y-axis range
ax.set_ylim([0, 45])

# add annotation for a specific item
ax.annotate('toyota corolla',
			xy=(19,33.9),
			xytext=(21, 35),
			arrowprops=dict(facecolor='black', shrink=.05)
)
```
