# web scraping

## data sourcing via web

```python
# import libraries
from bs4 import BeautifulSoup
import urllib.request
from IPython.display import HTML
import re
```

```python
# get data
r = urllib.request.urlopen('https://analytics.usa.gov/').read()
soup = BeautifulSoup(r)
type(soup)
```

```python
# display example data
print(soup.prettify()[:100])
```

```python
# print all link contained in the html data
for link in soup.find_all('a'):
	# print(link)
	print(link.get('href'))
```

```python
# display entire body of text
print(soup.get_text())
```

```python
# grep only valid links from hyperlink attribute
for link in soup.find_all('a', attrs={'href' : re.compile('^https')}): # starts w/ https
	print(link.get('href'))
```

```python
# store data in a file
# open w/ write mode
file = open('linkedin-data/parsed_data.txt', 'w') # w = over-write, a = append

# write links to the file
for link in soup.find_all('a', attrs={'href' : re.compile('^https')}):
	file.write(str(link.get_text()).strip() + ' : '+ link.get('href') + '\n')

# close the file
file.close()

# re-open as read mode to see the result
file = open('linkedin-data/parsed_data.txt', 'r')

# observe flush behavior
print('--- before flushing ---')
content = file.read()
print(content)

file.flush() # clears the internal buffer of the file, nothing to do w/ the content of the file

print('--- after flushing ---')
content = file.read()
print(content)

file.close()

# re-open to recheck the content of the file
file = open('linkedin-data/parsed_data.txt', 'r')

# print the result
print('--- after re-open ---')
content = file.read()
print(content)
```
